```python
def tc_ephem(time_span):
    start_time = time_span[0]
    time_step  = time_span[2]
    stop_time  = time_span[1]+time_step
    time       = np.arange(start_time,stop_time,time_step)
    num_step   = len(time)
    eph        = np.zeros((num_step,10))
    eph[:,0]   = time[:]
    eph[:,1]   = time[:]
    eph[:,2]   = time[:]*time[:]
    eph[:,3]   = time[:]*time[:]*time[:]
    eph[:,4]   = np.ones(num_step)
    eph[:,5]   = 2.0*time[:]
    eph[:,6]   = 3.0*time[:]*time[:]
    eph[:,7]   = np.zeros(num_step)
    eph[:,8]   = 2.0*np.ones(num_step)
    eph[:,9]   = 6.0*time[:]
    
    return eph
```

```python
def VNB(eph):
    num_step     = len(eph[:])
    moving_frame = np.zeros((num_step,9))
    for i in range(num_step):
        r    = eph[i,1:4]
        v    = eph[i,4:7]
        vmag = np.sqrt(v.dot(v)) 
        a    = eph[i,7:10]
        V    = v/vmag
        L    = np.cross(r,v)
        Ldot = np.cross(r,a)
        Lmag = np.sqrt(L.dot(L)) 
        N    = L/Lmag
        B    = np.cross(V,N)
        dV = a/vmag - (v.dot(a)/vmag**3)*v
        dN = Ldot/Lmag - L*(L.dot(Ldot)/Lmag**3)
        dB = np.cross(dV,N) + np.cross(V,dN)
        record = [V.dot(dV),V.dot(dN),V.dot(dB),N.dot(dV),N.dot(dN),N.dot(dB),B.dot(dV),B.dot(dN),B.dot(dB)]
        moving_frame[i,:] = np.array(record)
        
    return moving_frame    
```

```python
def RIC(eph):
    num_step     = len(eph[:])
    moving_frame = np.zeros((num_step,9))
    for i in range(num_step):
        r    = eph[i,1:4]
        rmag = np.sqrt(r.dot(r))
        v    = eph[i,4:7]
        vmag = np.sqrt(v.dot(v)) 
        a    = eph[i,7:10]
        R    = r/rmag
        L    = np.cross(r,v)
        Ldot = np.cross(r,a)
        Lmag = np.sqrt(L.dot(L)) 
        C    = L/Lmag
        I    = np.cross(C,R)
        dR = v/rmag - (r.dot(v)/rmag**3)*r
        dC = Ldot/Lmag - L*(L.dot(Ldot)/Lmag**3)
        dI = np.cross(dC,R) + np.cross(C,dR)
        record = [R.dot(dR),R.dot(dI),R.dot(dC),I.dot(dR),I.dot(dI),I.dot(dC),C.dot(dR),C.dot(dI),C.dot(dC)]
        moving_frame[i,:] = np.array(record)
        
    return moving_frame
```

```python
time_span = [-2,2,0.1]
```

```python
eph    = tc_ephem(time_span)
vbn_mf = VNB(eph)  
ric_mf = RIC(eph)
vbn_mf[np.abs(vbn_mf) < 2.0e-15] = 0 
ric_mf[np.abs(ric_mf) < 2.0e-15] = 0 
```
