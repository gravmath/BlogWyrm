```python
import matplotlib        as mpl
import matplotlib.pyplot as plt
import numpy             as np
import pickle
```

```python
data_files = ['data_128.pkl',
              'data_64.pkl',
              'data_32.pkl',
              'data_16.pkl',]
```

```python
data = []
for df in data_files:
    with open(df,'rb') as file:
        data.append(pickle.load(file))
```

```python
N = 45
sizes    = [16,32,64,128]
```

```python
N = 45
fig_vlo, axes_vlo  = plt.subplots(2,2,figsize=(6,6))
mats     = [data[3]['vlo']['C'][N][3],
            data[2]['vlo']['C'][N][3],
            data[1]['vlo']['C'][N][3],
            data[0]['vlo']['C'][N][3]]
counter = 0 
for i in range(2):
    for j in range(2):
        axes_vlo[i,j].imshow(mats[i*2+j])
        axes_vlo[i,j].set_title('Size = %s' % sizes[2*i+j])
        axes_vlo[i,j].axis('off')
        
big_ax = fig_vlo.add_subplot(1,1,1)
big_ax.axis('off')
big_ax.set_title('Probability p = 0.4\n',fontsize=16)
fig_vlo.savefig('UTH_03(Mar)_2024_Percolation p=0.4.png',dpi=135)
```

```python
N = 0
fig_ns,axes_ns = plt.subplots(2,3,figsize=(12,8))
counter = 0
ns = [data[N]['vlo']['A'],
      data[N]['lo'] ['A'],
      data[N]['c']  ['A'],
      data[N]['hi'] ['A'],
      data[N]['vhi']['A'],]
probs = [0.4,0.5,0.5927,0.7,0.8]
for i in range(2):
    for j in range(3):
        if counter == 5:
            break
        axes_ns[i,j].loglog(ns[2*i+j][0],ns[2*i+j][1],'k.')
        axes_ns[i,j].set_title('Probability = %s' % probs[2*i+j])
        axes_ns[i,j].set_xlim([1,1e4])
        axes_ns[i,j].set_ylim([1e-6,1e-2])
        #axes_ns[i,j].axis('off')
        
axes_ns[1,2].loglog(data[N]['vhi']['A'][0],data[0]['vhi']['A'][1],'b.',label='p=0.8')
axes_ns[1,2].loglog(data[N]['hi']['A'][0], data[0]['hi']['A'][1],'c.',label='p=0.7')
axes_ns[1,2].loglog(data[N]['c']['A'][0],  data[0]['c']['A'][1],'g.',label='p=p_c')
axes_ns[1,2].loglog(data[N]['lo']['A'][0], data[0]['lo']['A'][1],'r.',label='p=0.5')
axes_ns[1,2].loglog(data[N]['vlo']['A'][0],data[0]['vlo']['A'][1],'k.',label='p=0.4')
axes_ns[1,2].set_xlim([1,1e4])
axes_ns[1,2].set_ylim([1e-6,1e-2])
axes_ns[1,2].legend()
```
