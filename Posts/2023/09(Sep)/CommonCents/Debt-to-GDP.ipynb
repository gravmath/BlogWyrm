```python
import matplotlib.pyplot as plt
import numpy             as np
import pandas            as pd
```

```python
d2g = pd.read_csv('debt_to_gdp_percentage.csv')
```

```python
d2g
```

```python
fig = plt.figure(figsize=(14,7))
ax  = fig.add_subplot(1,1,1)
ax.set_facecolor("#aaaaaa")
ax.plot(d2g['Date'],d2g['D2G'],color="#f5f5dc")
ax.fill_between(d2g['Date'],d2g['D2G'],y2=0,color="#f5f5dc")
ax.grid('on',color="#888")
ax.set_xlabel('Year',fontsize=16)
ax.set_ylabel('Debt to GDP Percentage',fontsize=16)
ax.set_ylim([0,140])
#ax.set_xlim([1790,1920])
#fig.savefig('D2G_1790-to_2022.png')
```
