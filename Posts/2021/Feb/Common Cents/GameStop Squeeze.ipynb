```python
import datetime   as dt
import matplotlib as mpl
import numpy      as np
import pandas     as pd
import scipy      as sp
import matplotlib.pyplot as plt
```

```python
fn = 'd:github/BlogWyrm/Posts/2021/Feb/Common Cents/GME_Stock_Price_Data_5year.csv'
GME_df = pd.read_csv(fn,parse_dates=[0],infer_datetime_format=True)
```

```python
GME_df
```

```python
GME_df['Date'][0]
```

```python
GME_df['Range'] = GME_df['High'] - GME_df['Low']
GME_df['Ave']   = (GME_df['High'] + GME_df['Low'])/2
GME_df['Exchange'] = GME_df['Ave']*GME_df['Volume']
```

```python
billion = 1e9
plt.plot(GME_df['Date'],GME_df['Exchange']/billion,'r-')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Billion Dollars',fontsize=12)
plt.savefig('d:/github/BlogWyrm/Posts/2021/Feb/Common Cents/movement_linear.png')
```

```python
billion = 1e9
x_min = mpl.dates.date2num(dt.date(2020,3,10))
x_max = mpl.dates.date2num(dt.date(2021,3,10))
plt.semilogy(GME_df['Date'],GME_df['Exchange']/billion,'r-')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Billion Dollars',fontsize=12)
plt.xlim([x_min,x_max])
plt.savefig('d:/github/BlogWyrm/Posts/2021/Feb/Common Cents/movement_zoom_log.png')
```

```python
plt.plot(GME_df['Date'],GME_df['Volume']/1e6,'r-')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Million Shares',fontsize=12)
plt.savefig('d:/github/BlogWyrm/Posts/2021/Feb/Common Cents/volume_5yr.png')
```

```python
plt.plot(GME_df['Date'],GME_df['Ave'],'r-')
plt.xlabel('Date',fontsize=12)
plt.ylabel('Average Daily Share Price (dollars)',fontsize=12)
plt.savefig('d:/github/BlogWyrm/Posts/2021/Feb/Common Cents/ave_price_5yr.png')
```
