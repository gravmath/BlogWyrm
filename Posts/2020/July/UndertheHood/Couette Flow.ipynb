```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def time_varying_Couette_Acheson(t,y_arr,order,U,w,nu):
    from numpy import pi as pi
    
    soln = U*(1-y_arr/w)  #steady solution
    
    for n in range(1,order+1):
        coeff       = -2.0*U/(n*pi)
        time_piece  = np.exp(-(n*pi/w)**2 * nu * t)
        space_piece = np.sin(n*pi/w*y_arr)
        soln += coeff*time_piece*space_piece
    
    return soln    
```

```python
def time_varying_Couette(t,y_arr,order,U,w,nu):
    from numpy import pi as pi
    
    soln = U*y_arr/w  #steady solution
    
    for n in range(1,order+1):
        coeff       = (-1)**n * 2.0*U/(n*pi)
        time_piece  = np.exp(-(n*pi/w)**2 * nu * t)
        space_piece = np.sin(n*pi/w*y_arr)
        soln += coeff*time_piece*space_piece
    
    return soln    
```

```python
U = 50
w = 5
nu = 0.8926
t = 0
N = 51
order = 1000
y = np.linspace(0,w,N)
```

```python
soln = time_varying_Couette_Acheson(t,y,order,U,w,nu)
```

```python
plt.plot(soln,y)
```

```python
omega = 10
k = np.sqrt(omega/nu/2)
U = 50
w = 5
N = 91
y = np.linspace(0,w,N)
```

```python
def shaky_plate_flow(t,y_arr,U,omega,k):
    return U*np.exp(-k*y_arr)*np.cos(omega*t - k*y_arr)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10))
counter = 0
for t in np.arange(0,3*2.0*np.pi/omega,0.002):
    soln = shaky_plate_flow(t,y,U,omega,k)
    ax.set_xlabel('Flow velocity (mm/s)',fontsize=20)
    ax.set_ylabel('Distance from bottom plate (mm)',fontsize=20)
    ax.set_xlim([-55,55])
    ax.set_title('t = {:1.3f}'.format(t))
    ax.plot(soln,y)
    fig.savefig('c:/users/cschiff/Documents/shaky_t{:04d}.jpg'.format(counter))
    counter += 1
    plt.cla()
print('all done!')
```
