```python
import numpy             as np
import numpy.random      as random
import random            as rnd
import scipy             as sp
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,3.5,0.5)
y = np.arange(-3,3.5,0.5)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
def flow(x,y,alpha):
    U =  alpha*x
    V = -alpha*y
    
    return U, V
```

```python
U, V = flow(xx,yy,1)
```

```python
fig_q = plt.figure(figsize=(10,10))
ax_q  = fig_q.add_subplot(1,1,1)
ax_q.quiver(xx,yy,U,V,np.hypot(U,V),scale=50)
ax_q.set_xlabel('X')
ax_q.set_ylabel('Y')
fig_q.savefig('c:/Users/byecs/Documents/UTH_04_24_2020-fluid-flow.png')
```

```python
def flow2(x,y,t):
    U = x**(0)
    V = y**(0)*t
    
    return U, V
```

```python
U0, V0 = flow2(xx,yy,0)
U1, V1 = flow2(xx,yy,1)
```

```python
fig_Q = plt.figure(figsize=(10,5))
ax_q0 = fig_Q.add_subplot(1,2,1)
ax_q1 = fig_Q.add_subplot(1,2,2)
ax_q0.quiver(xx,yy,U0,V0,np.hypot(U0,V0),scale=20)
ax_q0.set_xlabel('X')
ax_q0.set_ylabel('Y')
ax_q0.set_title('t=0')
ax_q1.quiver(xx,yy,U1,V1,np.hypot(U1,V1),scale=20)
ax_q1.set_xlabel('X')
ax_q1.set_ylabel('Y')
ax_q1.set_title('t=1')
fig_Q.savefig('c:/Users/byecs/Documents/UTH_04_24_2020-fluid-flow-unsteady.png')
```
