```python
import matplotlib as mpl
mpl.use('agg')
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def no_match_prob(n):
    prod = 1.0
    for i in range(1,n):
        prod = prod * (365.0 - i)/365.0
        
    return prod
```

```python
1.0 - no_match_prob(23)
```

```python
num_people = np.array(range(1,60))
```

```python
match_prob = (np.ones(len(num_people)) - map(no_match_prob,num_people))*100.0
```

```python
plt.plot(num_people,match_prob,color = '#ff6666',marker='o')
plt.grid('on')
plt.xlabel('Number of People')
plt.ylabel('Probability of a Match')
```
