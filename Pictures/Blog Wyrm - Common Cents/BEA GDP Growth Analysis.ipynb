```python
GDP_growth    = np.zeros(len(year))
GDP_growth[0] = np.nan
for _ in range(1,len(year)):
    GDP_growth[_] = ( GDP_2009[_] - GDP_2009[_-1] )/GDP_2009[_-1]*100
GDP_growth = np.ma.masked_invalid(GDP_growth)
```

```python
fig2 = plt.figure()
ax2  = fig2.add_subplot(1,1,1)
ax2.set_axis_bgcolor('#eeeeee')
ax2.plot(year,GDP_growth,color='#d3d3ba',marker='o')
ax2.set_xlabel('Year')
ax2.set_ylabel('GDP Percent Growth')
ax2.axhline(3.34,color='#222222')
ax2.grid('on')
fig2.savefig('c:/Users/Conrad/Documents/GitHub/BlogWyrm/Pictures/Blog Wyrm - Common Cents/GDP_growth.png')
```
